```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://dzen.ru/a/YZ3txZxykGdEJNjz')
soup = BeautifulSoup(r.content, features='html.parser')
```

```python
news = soup.find_all('item')
```

```python
print(len(news))
print(news[0])
```

```python
print(news[0].title.text)
print(news[0].description.text)
print(news[0].guid.text)
```

```python
f = open('news.txt', 'w')
newline = '\n'
for item in news:
  print(item.title.text)
  item_link = item.guid.text
  r2 = requests.get(item_link)
  soup2 = BeautifulSoup(r2.content, features='html.parser')
  paragraphs = soup2.find_all('p')
  res1 = f'{item.title.text} \n \n'
  f.write("".join(res1))

  for paragr in paragraphs:
    print(paragr.text)
    res2 = f'{paragr.text} \n'
    f.write(res2)
  res3 = f'{item_link} \n ================================================ \n'
  f.write("".join(res3))
  #print(item.guid.text)
  #if item.category.text == 'В России':
   # print(item.title.text)
    #print(item.description.text)
    #print(item.guid.text)
    #print('=================================== \n')
f .close()
```
